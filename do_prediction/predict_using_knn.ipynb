```python
import pandas as pd
import joblib

# LOAD MODEL + SCALER

knn = joblib.load("../trained_models/knn_model.pkl")
scaler = joblib.load("../trained_models/knn_scaler.pkl")

# Load processed dataset to get correct columns
df_processed = pd.read_csv("../processed_datasets/knn_processed.csv")
X_columns = df_processed.drop("Churn_Yes", axis=1).columns

print("Model loaded successfully!")

# FUNCTION: RAW INPUT → PREDICTION

def predict_churn(raw_dict):
    # Convert raw input to DataFrame
    df_raw = pd.DataFrame([raw_dict])
    
    # Apply One-Hot Encoding
    df_encoded = pd.get_dummies(df_raw)
    
    # Match columns with training data
    df_encoded = df_encoded.reindex(columns=X_columns, fill_value=0)
    
    # Scale input
    df_scaled = scaler.transform(df_encoded)
    
    # Predict
    prediction = knn.predict(df_scaled)[0]
    
    # Output result
    if prediction == 1:
        return "Customer will CHURN"
    else:
        return "Customer will STAY"


# TEST WITH RAW INPUT

# Example input (same format as original dataset)
raw_customer = {
    "gender": "Male",
    "SeniorCitizen": 0,
    "Partner": "Yes",
    "Dependents": "No",
    "tenure": 3,
    "PhoneService": "Yes",
    "MultipleLines": "No",
    "InternetService": "Fiber optic",
    "OnlineSecurity": "No",
    "OnlineBackup": "Yes",
    "DeviceProtection": "No",
    "TechSupport": "No",
    "StreamingTV": "Yes",
    "StreamingMovies": "Yes",
    "Contract": "Month-to-month",
    "PaperlessBilling": "Yes",
    "PaymentMethod": "Electronic check",
    "MonthlyCharges": 85,
    "TotalCharges": 255
}

# Run prediction
result = predict_churn(raw_customer)

print("\nPrediction Result:")
print(result)
```
