```python
# =============================================
# WILL CHURN PREDICTION - Using Saved Random Forest Model
# =============================================

from google.colab import drive
import pandas as pd
import numpy as np
import joblib

# ============================
# 1. MOUNT GOOGLE DRIVE
# ============================
drive.mount('/content/drive')

# ============================
# 2. LOAD THE SAVED MODEL
# ============================
model_path = '/content/drive/MyDrive/random_forest_churn_model.pkl'   # ← Update path if different

rf_model = joblib.load(model_path)
print("✅ Random Forest model loaded successfully from Google Drive!")

# ============================
# 3. CREATE A "WILL CHURN" CUSTOMER PROFILE
# ============================

# This customer profile is designed with high-churn risk characteristics:
# - Short tenure, Month-to-month contract, Fiber optic, high monthly charges, etc.

new_customer = {
    'gender': 'Female',
    'SeniorCitizen': 1,                    # Senior citizen (higher risk)
    'Partner': 'No',
    'Dependents': 'No',
    'tenure': 3,                           # Very short tenure
    'PhoneService': 'Yes',
    'MultipleLines': 'Yes',
    'InternetService': 'Fiber optic',      # Expensive service
    'OnlineSecurity': 'No',
    'OnlineBackup': 'No',
    'DeviceProtection': 'No',
    'TechSupport': 'No',                   # No support = higher churn
    'StreamingTV': 'Yes',
    'StreamingMovies': 'Yes',
    'Contract': 'Month-to-month',          # Biggest churn driver
    'PaperlessBilling': 'Yes',
    'PaymentMethod': 'Electronic check',   # Often linked to churn
    'MonthlyCharges': 95.50,               # High monthly bill
    'TotalCharges': 286.50                 # Low total because new customer
}

# Convert to DataFrame (1 row)
df_new = pd.DataFrame([new_customer])

print("\n📋 Customer Profile for Prediction (High Churn Risk):")
print(df_new.T)   # Transposed for easy reading

# ============================
# 4. PREPROCESS THE NEW RECORD
# ============================

# Fix data types
df_new['TotalCharges'] = pd.to_numeric(df_new['TotalCharges'], errors='coerce')

# Encode categorical columns (same logic as training)
categorical_cols = df_new.select_dtypes(include=['object']).columns.tolist()

for col in categorical_cols:
    # Simple Label Encoding (for demo purposes)
    unique_values = df_new[col].unique()
    mapping = {val: idx for idx, val in enumerate(unique_values)}
    df_new[col] = df_new[col].map(mapping)

print("\n✅ Customer data preprocessed for prediction!")

# ============================
# 5. MAKE THE PREDICTION
# ============================

prediction = rf_model.predict(df_new)
prediction_proba = rf_model.predict_proba(df_new)[:, 1][0] * 100

churn_status = "WILL CHURN" if prediction[0] == 1 else "Will NOT Churn"

print("\n" + "="*70)
print("🎯 PREDICTION RESULT - WILL CHURN?")
print("="*70)
print(f"Prediction          : **{churn_status}**")
print(f"Churn Probability   : {prediction_proba:.2f}%")
print("="*70)

if prediction[0] == 1:
    print("🔴 This customer is predicted to churn with high probability.")
else:
    print("🟢 This customer is predicted to stay.")

# ============================
# 6. Optional: Show Top Important Features (for explanation)
# ============================
if hasattr(rf_model, 'feature_importances_'):
    # Note: This requires the feature names from training.
    # For simplicity, we skip detailed names here.
    print("\nTip: In your training notebook, you showed feature importance (Contract, tenure, MonthlyCharges, etc. are usually top drivers).")
```
