## PREDICTION ONLY - Using Saved Random Forest Model

```python
# =============================================
# NO CHURN PREDICTION (Low Risk Customer)
# =============================================

from google.colab import drive
import pandas as pd
import joblib

drive.mount('/content/drive')

# Load model and encoders
model_path = '/content/drive/MyDrive/random_forest_churn_model_smote.pkl'   # or without _smote
encoders_path = '/content/drive/MyDrive/label_encoders.pkl'

rf_model = joblib.load(model_path)
le_dict = joblib.load(encoders_path)

print("✅ Model and LabelEncoders loaded!")

# Low risk customer (expected: Will NOT Churn)
new_customer = {
    'gender': 'Male',
    'SeniorCitizen': 0,
    'Partner': 'Yes',
    'Dependents': 'Yes',
    'tenure': 48,
    'PhoneService': 'Yes',
    'MultipleLines': 'No',
    'InternetService': 'DSL',
    'OnlineSecurity': 'Yes',
    'OnlineBackup': 'Yes',
    'DeviceProtection': 'Yes',
    'TechSupport': 'Yes',
    'StreamingTV': 'No',
    'StreamingMovies': 'No',
    'Contract': 'Two year',
    'PaperlessBilling': 'No',
    'PaymentMethod': 'Credit card (automatic)',
    'MonthlyCharges': 65.50,
    'TotalCharges': 3120.0
}

df_new = pd.DataFrame([new_customer])

# Preprocess using SAVED encoders (CRITICAL!)
for col in le_dict:
    if col in df_new.columns:
        df_new[col] = le_dict[col].transform(df_new[col])

print("\n📋 Low Risk Customer Profile:")
print(df_new.T)

# Make prediction
prediction = rf_model.predict(df_new)[0]
proba = rf_model.predict_proba(df_new)[0][1] * 100

print("\n" + "="*65)
print("🎯 NO CHURN PREDICTION RESULT")
print("="*65)
print(f"Prediction          : {'Will Churn' if prediction == 1 else 'Will NOT Churn'}")
print(f"Churn Probability   : {proba:.2f}%")
print("="*65)
```
